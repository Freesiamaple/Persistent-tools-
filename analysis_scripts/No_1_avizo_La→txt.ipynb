```python
#csvファイルの名前 例)abcde.csv なら "abcde"
#Label-Analysis からhomcloudに入れる用のtxtを出す
name="111d4"
```

```python
import math
import numpy as np
import pandas as pd
DATA_DIR = "../data/model data/"
CACHE_DIR = "../data/cache/"
IMAGE_DIR = "../data/images/"
INVERSE_DIR = "../data/Inverse analysis/"
EXCEL_DIR = "../data/excel/"
PKL_DIR = "../data/pkl/"
```

```python
def convert_csv_to_pkl(filename):
    df = pd.read_csv(f'{EXCEL_DIR}{filename}.csv',header=1)
    df.to_pickle(f'{PKL_DIR}{filename}.pkl')

#以下 convert_csv_to_pkl(filename) でcsv読み込み

convert_csv_to_pkl(name)
```

```python
df=pd.read_pickle(f'{PKL_DIR}{name}.pkl')
df
```

```python
se=pd.Series(df['Volume3d (?m^3) '] ,dtype='float64')
def volume_to_radius2(x):
    return float((x*3/(4*math.pi))**(2/3))
df["radius2"]=se.map(volume_to_radius2)
df
```

```python
df.rename(columns={"BaryCenterX (?m) ": 'CenterX'}, inplace=True)
df.rename(columns={"BaryCenterY (?m) ": 'CenterY'}, inplace=True)
df.rename(columns={"BaryCenterZ (?m) ": 'CenterZ'}, inplace=True)
```

```python
df_2=df
```

```python
base_df=df_2[["CenterX","CenterY","CenterZ","radius2"]]
base_df
```

```python
base_df.to_csv(f'{DATA_DIR}{name}.txt', sep='\t',index=False,header=False)
base_df
```
