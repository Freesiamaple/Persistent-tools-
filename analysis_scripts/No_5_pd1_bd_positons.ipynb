```python
#xyzr^2のtxtから、b d bp1 bp2 dp1 dp2 dp3の形式のcsvに変換するコード(1行で書けるかも)
filename='111d4'
```

```python
EXCEL_DIR = "../data/excel/"
DATA_DIR = "../data/model data/"
CACHE_DIR = "../data/cache/"
```

```python
import homcloud.interface as hc  # HomCloudのインターフェス
import numpy as np
import matplotlib.pyplot as plt
import homcloud.paraview_interface as pv




#pdgm file
pointcloud = np.loadtxt('{}{}.txt'.format(DATA_DIR,filename))
pointcloud.shape
hc.PDList.from_alpha_filtration(pointcloud, save_to="{}{}.pdgm".format(CACHE_DIR,filename),save_boundary_map=True,weight=True,no_squared=True)

#pd1
pdlist = hc.PDList("{}{}.pdgm".format(CACHE_DIR,filename))
pd1 = pdlist.dth_diagram(1)
np.savetxt('{}{}_b_pd1.txt'.format(CACHE_DIR,filename),pd1.births,header='b')
np.savetxt('{}{}_d_pd1.txt'.format(CACHE_DIR,filename),pd1.deaths,header='d')

print("end")
```

```python
pd1.death_positions
print(pd1.birth_positions[0])
L=len(pd1.births)
print(pd1.births[0])
import pandas as pd
```

```python
B=pd.Series(list(pd1.births))
D=pd.Series(list(pd1.deaths))
B_P=pd.Series(list(pd1.birth_positions))
print(len(B_P[0]))
```

```python
B_P_name=["Birth_Position_1_X","Birth_Position_1_Y","Birth_Position_1_Z","Birth_Position_2_X","Birth_Position_2_Y","Birth_Position_2_Z"]
data_B_P_1_x=[]
data_B_P_1_y=[]
data_B_P_1_z=[]
data_B_P_2_x=[]
data_B_P_2_y=[]
data_B_P_2_z=[]
for i in range(L):
    data_B_P_1_x.append(B_P[i][0][0])
    data_B_P_1_y.append(B_P[i][0][1])
    data_B_P_1_z.append(B_P[i][0][2])
    data_B_P_2_x.append(B_P[i][1][0])
    data_B_P_2_y.append(B_P[i][1][1])
    data_B_P_2_z.append(B_P[i][1][2])
series_1_x = pd.Series(data=data_B_P_1_x)
series_1_y = pd.Series(data=data_B_P_1_y)
series_1_z = pd.Series(data=data_B_P_1_z)
series_2_x = pd.Series(data=data_B_P_2_x)
series_2_y = pd.Series(data=data_B_P_2_y)
series_2_z = pd.Series(data=data_B_P_2_z)
```

```python
df=pd.DataFrame(B)
df_=df.rename(columns={0:"Birth"},inplace=True)
df["Death"]=D
df.head()
```

```python
D_P=pd.Series(list(pd1.death_positions))
```

```python
df[B_P_name[0]]=series_1_x
df[B_P_name[1]]=series_1_y
df[B_P_name[2]]=series_1_z
df[B_P_name[3]]=series_2_x
df[B_P_name[4]]=series_2_y
df[B_P_name[5]]=series_2_z
df
```

```python
D_P_name=["Death_Position_1_X","Death_Position_1_Y","Death_Position_1_Z","Death_Position_2_X","Death_Position_2_Y","Death_Position_2_Z","Death_Position_3_X","Death_Position_3_Y","Death_Position_3_Z"]
data_D_P_1_x=[]
data_D_P_1_y=[]
data_D_P_1_z=[]
data_D_P_2_x=[]
data_D_P_2_y=[]
data_D_P_2_z=[]
data_D_P_3_x=[]
data_D_P_3_y=[]
data_D_P_3_z=[]

for i in range(L):
    data_D_P_1_x.append(D_P[i][0][0])
    data_D_P_1_y.append(D_P[i][0][1])
    data_D_P_1_z.append(D_P[i][0][2])
    data_D_P_2_x.append(D_P[i][1][0])
    data_D_P_2_y.append(D_P[i][1][1])
    data_D_P_2_z.append(D_P[i][1][2])
    data_D_P_3_x.append(D_P[i][2][0])
    data_D_P_3_y.append(D_P[i][2][1])
    data_D_P_3_z.append(D_P[i][2][2])
    
series_d_1_x = pd.Series(data=data_D_P_1_x)
series_d_1_y = pd.Series(data=data_D_P_1_y)
series_d_1_z = pd.Series(data=data_D_P_1_z)
series_d_2_x = pd.Series(data=data_D_P_2_x)
series_d_2_y = pd.Series(data=data_D_P_2_y)
series_d_2_z = pd.Series(data=data_D_P_2_z)
series_d_3_x = pd.Series(data=data_D_P_3_x)
series_d_3_y = pd.Series(data=data_D_P_3_y)
series_d_3_z = pd.Series(data=data_D_P_3_z)
df[D_P_name[0]]=series_d_1_x
df[D_P_name[1]]=series_d_1_y
df[D_P_name[2]]=series_d_1_z
df[D_P_name[3]]=series_d_2_x
df[D_P_name[4]]=series_d_2_y
df[D_P_name[5]]=series_d_2_z
df[D_P_name[6]]=series_d_3_x
df[D_P_name[7]]=series_d_3_y
df[D_P_name[8]]=series_d_3_z
df
```
