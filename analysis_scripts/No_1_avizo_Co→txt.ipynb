```python
#csvファイルの名前 例)abcde.csv なら "abcde"
name="load0b"
```

```python
import math
import numpy as np
import pandas as pd
```

```python
DATA_DIR = "../data/model data/"
CACHE_DIR = "../data/cache/"
IMAGE_DIR = "../data/images/"
INVERSE_DIR = "../data/Inverse analysis/"
EXCEL_DIR = "../data/excel/"
PKL_DIR = "../data/pkl/"
```

```python
def convert_csv_to_pkl(filename):
    df = pd.read_csv(f'{EXCEL_DIR}{filename}.csv',header=1)
    df.to_pickle(f'{PKL_DIR}{filename}.pkl')

#以下 convert_csv_to_pkl(filename) でcsv読み込み

convert_csv_to_pkl(name)
```

```python
original_df=pd.read_pickle(f'{PKL_DIR}{name}.pkl')
original_df
```

```python
original_df['Volume [?m^3]']
se=pd.Series(original_df['Volume [?m^3]'] ,dtype='float64')
def volume_to_radius2(x):
    return str((x*3/(4*math.pi))**(2/3))
```

```python
se.map(volume_to_radius2)
```

```python
original_df["radius2"]=se.map(volume_to_radius2)
original_df
```

```python
base_df=original_df[["CenterX [?m]","CenterY [?m]","CenterZ [?m]","radius2"]]
```

```python
base_df.to_csv(f'{DATA_DIR}{name}.txt', sep='\t',index=False,header=False)
base_df
```
