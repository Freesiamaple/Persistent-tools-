```python
Sphericity_df=pd.read_pickle(f'{PKL_DIR}{named}.pkl')
Sphericity_df
```

```python
se=pd.Series(Sphericity_df['Volume3d (?m^3) '] ,dtype='float64')
def volume_to_surface(x):
    return math.pi**(1/3)*(6*x)**(2/3)
```

```python
se.map(volume_to_surface)
```

```python
Sphericity_df["Surface"]=se.map(volume_to_surface)
```

```python
Sphericity_df['Sphericity']=Sphericity_df["Surface"]/Sphericity_df["Area3d (?m^2) "]
```
